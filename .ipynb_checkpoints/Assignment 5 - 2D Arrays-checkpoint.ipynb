Convert 1D Array Into 2D Array

You are given a **0-indexed** 1-dimensional (1D) integer array original, and two integers, m and n. You are tasked with creating a 2-dimensional (2D) array with Â m rows and n columns using **all** the elements from original.

The elements from indices 0 to n - 1 (**inclusive**) of original should form the first row of the constructed 2D array, the elements from indices n to 2 * n - 1 (**inclusive**) should form the second row of the constructed 2D array, and so on.

Return *an* m x n *2D array constructed according to the above procedure, or an empty 2D array if it is impossible*.

**Example 1:**

**Input:** original = [1,2,3,4], m = 2, n = 2

**Output:** [[1,2],[3,4]]

**Explanation:** The constructed 2D array should contain 2 rows and 2 columns.

The first group of n=2 elements in original, [1,2], becomes the first row in the constructed 2D array.

The second group of n=2 elements in original, [3,4], becomes the second row in the constructed 2D array.

**Ans:-**
To convert a 1D array into a 2D array according to the given procedure, we can follow the following steps:

1. Check if the total number of elements in the original array is equal to m * n. If not, return an empty 2D array.

2. Create an empty 2D array with m rows and n columns.

3. Iterate over the elements of the original array. For each element, calculate the corresponding row and column indices in the 2D array.

4. Assign the element to the calculated row and column in the 2D array.

5. Return the constructed 2D array.

````python
def construct2DArray(original, m, n):
    if len(original) != m * n:
        return []

    result = [[0] * n for _ in range(m)]
    for i, num in enumerate(original):
        row = i // n
        col = i % n
        result[row][col] = num

    return result
````

ðŸ’¡ **Question 2**
You have n coins and you want to build a staircase with these coins. The staircase consists of k rows where the ith row has exactly i coins. The last row of the staircase **may be** incomplete.

Given the integer n, return *the number of **complete rows** of the staircase you will build*.

**Example 1:**
```
**Input:** n = 5

**Output:** 2

**Explanation:** Because the 3rd row is incomplete, we return 2.
```
</aside>

**Ans:-**
To solve this problem, we can use a simple approach. We start with the first row and keep adding rows until we run out of coins or cannot complete the next row.

Here's the step-by-step algorithm:

1. Initialize a variable `completeRows` to 0. This variable will keep track of the number of complete rows.

2. Initialize a variable `coinsRequired` to 1. This variable represents the number of coins required to complete the next row.

3. While `n` is greater than or equal to `coinsRequired`:
   - Increment `completeRows` by 1.
   - Subtract `coinsRequired` from `n`.
   - Increment `coinsRequired` by 1.

4. Return the value of `completeRows` as the result.

````python
def numCompleteRows(n):
    completeRows = 0
    coinsRequired = 1

    while n >= coinsRequired:
        completeRows += 1
        n -= coinsRequired
        coinsRequired += 1

    return completeRows
````

````python
numCompleteRows(5)
````

ðŸ’¡ **Question 3**
Given an integer array nums sorted in **non-decreasing** order, return *an array of **the squares of each number** sorted in non-decreasing order*.

**Example 1:**

Input: nums = [-4,-1,0,3,10]

Output: [0,1,9,16,100]

**Explanation:** After squaring, the array becomes [16,1,0,9,100].
After sorting, it becomes [0,1,9,16,100]

</aside>

**Ans:-**
To solve this problem, we can follow these steps:

1. Create an empty result list to store the squares of the numbers.
2. Iterate through each number in the input `nums` array.
3. Square each number and append it to the result list.
4. Sort the result list in non-decreasing order.
5. Return the sorted result list.

````python
def sortedSquares(nums):
    result = []

    for num in nums:
        square = num * num
        result.append(square)

    result.sort()
    return result
````

````python
nums = [-4,-1,0,3,10]
sortedSquares(nums)
````

ðŸ’¡ **Question 4**

Given two **0-indexed** integer arrays nums1 and nums2, return *a list* answer *of size* 2 *where:*

- answer[0] *is a list of all **distinct** integers in* nums1 *which are **not** present in* nums2*.*
- answer[1] *is a list of all **distinct** integers in* nums2 *which are **not** present in* nums1.

**Note** that the integers in the lists may be returned in **any** order.

**Example 1:**

**Input:** nums1 = [1,2,3], nums2 = [2,4,6]

**Output:** [[1,3],[4,6]]

**Explanation:**

For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas nums1[0] = 1 and nums1[2] = 3 are not present in nums2. Therefore, answer[0] = [1,3].

For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. Therefore, answer[1] = [4,6].

</aside>

**Ans:-**
To find the distinct integers in `nums1` that are not present in `nums2`, and vice versa, we can follow these steps:

1. Convert `nums1` and `nums2` to sets to remove duplicate elements.
2. Find the set difference between `nums1` and `nums2` to get the distinct elements in `nums1` that are not present in `nums2`.
3. Find the set difference between `nums2` and `nums1` to get the distinct elements in `nums2` that are not present in `nums1`.
4. Convert the resulting sets back to lists.
5. Return a list `[answer1, answer2]` where `answer1` is the list of distinct elements from `nums1` not present in `nums2`, and `answer2` is the list of distinct elements from `nums2` not present in `nums1`.

````python
def findDisjoint(nums1, nums2):
    set1 = set(nums1)
    set2 = set(nums2)

    answer1 = list(set1 - set2)
    answer2 = list(set2 - set1)

    return [answer1, answer2]
````

````python
nums1 = [1, 2, 3]
nums2 = [2, 4, 6]
result = findDisjoint(nums1, nums2)
print(result)
````

ðŸ’¡ **Question 5**

Given two integer arrays arr1 and arr2, and the integer d, *return the distance value between the two arrays*.

The distance value is defined as the number of elements arr1[i] such that there is not any element arr2[j] where |arr1[i]-arr2[j]| <= d.

**Example 1:**

**Input:** arr1 = [4,5,8], arr2 = [10,9,1,8], d = 2

**Output:** 2

**Explanation:**

For arr1[0]=4 we have:

|4-10|=6 > d=2

|4-9|=5 > d=2

|4-1|=3 > d=2

|4-8|=4 > d=2

For arr1[1]=5 we have:

|5-10|=5 > d=2

|5-9|=4 > d=2

|5-1|=4 > d=2

|5-8|=3 > d=2

For arr1[2]=8 we have:

**|8-10|=2 <= d=2**

**|8-9|=1 <= d=2**

|8-1|=7 > d=2

**|8-8|=0 <= d=2**

</aside>

**Ans:-**
To find the distance value between two arrays `arr1` and `arr2`, we can follow these steps:

1. Initialize a variable `distance` to 0 to keep track of the distance value.

2. Iterate over each element `num1` in `arr1`.

3. For each `num1`, iterate over each element `num2` in `arr2`.

4. Check if the absolute difference between `num1` and `num2` is greater than `d`. If it is, continue to the next `num2`.

5. If no `num2` is found such that the absolute difference is less than or equal to `d`, increment the `distance` by 1.

6. Return the final value of `distance`.

````python
def findDistanceValue(arr1, arr2, d):
    distance = 0
    for num1 in arr1:
        for num2 in arr2:
            if abs(num1 - num2) <= d:
                break
        else:
            distance += 1

    return distance
````

````python
arr1 = [4, 5, 8]
arr2 = [10, 9, 1, 8]
d = 2
result = findDistanceValue(arr1, arr2, d)
print(result)
````

ðŸ’¡ **Question 6**

Given an integer array nums of length n where all the integers of nums are in the range [1, n] and each integer appears **once** or **twice**, return *an array of all the integers that appears **twice***.

You must write an algorithm that runs in O(n) time and uses only constant extra space.

**Example 1:**

**Input:** nums = [4,3,2,7,8,2,3,1]

**Output:**

[2,3]

</aside>

**Ans:-**
To find all the integers that appear twice in the given array `nums`, while satisfying the constraints of O(n) time complexity and constant extra space, we can use the following approach:

1. Initialize an empty list called `result` to store the integers that appear twice.

2. Iterate over each element `num` in `nums`.

3. Get the absolute value of `num` and subtract 1 to obtain the index.

4. If the value at the obtained index is positive, it means that this is the first time we have encountered this number. Mark the value at the index as negative to indicate that we have seen it once.

5. If the value at the obtained index is negative, it means we have already encountered this number before. Add the absolute value of `num` to the `result` list.

6. Finally, return the `result` list containing all the integers that appear twice.

````python
def findDuplicates(nums):
    result = []
    for num in nums:
        index = abs(num) - 1
        if nums[index] > 0:
            nums[index] = -nums[index]
        else:
            result.append(abs(num))
    return result
````

````python
nums = [4, 3, 2, 7, 8, 2, 3, 1]
result = findDuplicates(nums)
print(result)
````

ðŸ’¡ **Question 7**

Suppose an array of length n sorted in ascending order is **rotated** between 1 and n times. For example, the array nums = [0,1,2,4,5,6,7] might become:

- [4,5,6,7,0,1,2] if it was rotated 4 times.
- [0,1,2,4,5,6,7] if it was rotated 7 times.

Notice that **rotating** an array [a[0], a[1], a[2], ..., a[n-1]] 1 time results in the array [a[n-1], a[0], a[1], a[2], ..., a[n-2]].

Given the sorted rotated array nums of **unique** elements, return *the minimum element of this array*.

You must write an algorithm that runs in O(log n) time.

**Example 1:**

**Input:** nums = [3,4,5,1,2]

**Output:** 1

**Explanation:**

The original array was [1,2,3,4,5] rotated 3 times.

</aside>

**Ans:-**
To find the minimum element in a sorted rotated array `nums` of unique elements, while satisfying the constraint of O(log n) time complexity, we can use a modified version of the binary search algorithm. Here's the approach:

1. Initialize two pointers, `left` and `right`, pointing to the start and end of the array respectively.

2. Perform a binary search while `left` is less than `right`.

3. Calculate the middle element index as `mid = (left + right) // 2`.

4. Check if `nums[mid]` is greater than `nums[right]`. If it is, it means the minimum element is on the right side of `mid`. Update `left = mid + 1`.

5. Otherwise, the minimum element is on the left side of `mid` or it could be the element at `mid` itself. Update `right = mid`.

6. Repeat steps 3-5 until `left` is no longer less than `right`.

7. The minimum element will be at index `left` (or `right`, both will be the same) in the array `nums`.

8. Return `nums[left]` (or `nums[right]`) as the minimum element.

````python
def findMin(nums):
    left = 0
    right = len(nums) - 1

    while left < right:
        mid = (left + right) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]   #here left or write both are correct
````

````python
nums = [3, 4, 5, 1, 2]
result = findMin(nums)
print(result)
````

````python
findMin([4,5,1,2,3])
````

ðŸ’¡ **Question 8**

An integer array original is transformed into a **doubled** array changed by appending **twice the value** of every element in original, and then randomly **shuffling** the resulting array.

Given an array changed, return original *if* changed *is a **doubled** array. If* changed *is not a **doubled** array, return an empty array. The elements in* original *may be returned in **any** order*.

**Example 1:**

**Input:** changed = [1,3,4,2,6,8]

**Output:** [1,3,4]

**Explanation:** One possible original array could be [1,3,4]:

- Twice the value of 1 is 1 * 2 = 2.
- Twice the value of 3 is 3 * 2 = 6.
- Twice the value of 4 is 4 * 2 = 8.

Other original arrays could be [4,3,1] or [3,1,4].

</aside>

**Ans:-**
To determine the original array from a transformed array `changed`, you can follow these steps:

1. Create a dictionary called `count` to keep track of the count of each element in `changed`.

2. Iterate over each element `num` in `changed` and update the count in the `count` dictionary.

3. Sort the unique elements in `changed` in ascending order.

4. Iterate over each element `num` in the sorted unique elements.

5. Check if the count of `num` in `count` is 0. If it is, continue to the next element.

6. Check if the count of `2 * num` in `count` is 0 or the count of `num` is less than the count of `2 * num`. If either condition is true, return an empty array since `changed` is not a doubled array.

7. Subtract the count of `2 * num` from the count of `num` in `count`.

8. Return a list of elements from the `count` dictionary where the value is greater than 0.

````python
from collections import Counter

def findOriginalArray(changed):
    count = Counter(changed)

    if len(changed) % 2 != 0:
        return []

    original = []

    for num in changed:
        if count[num] == 0:
            continue

        double = 2 * num

        if count[double] == 0:
            return []

        original.append(num)
        count[num] -= 1
        count[double] -= 1

    return original
````

````python
changed = [1, 3, 4, 2, 6, 8]
result = findOriginalArray(changed)
print(result)
````
